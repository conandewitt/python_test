```python
import threading
from IPython.display import display
import ipywidgets as widgets
import time
import numpy as np

from bokeh.plotting import figure,show
from bokeh.io import output_notebook, push_notebook
output_notebook()
```

```python
global interrupt
global data
```

```python
interrupt=False
```

```python
button=widgets.Button(description='start data logging')
display(button)

p1=figure(plot_width=650,plot_height=350)
t=show(p1,notebook_handle=True)

def when_clicked(b):
    def get_data():
        global interrupt
        global data
        j=0
        while not interrupt:     
                data=np.random.randint(1,100)
                r1=p1.circle(j,data,size=7)
                push_notebook(handle=t)
                j+=1
                time.sleep(0.5)


    thread = threading.Thread(target=get_data)
    thread.start()

button.on_click(when_clicked)
```

```python
button2=widgets.Button(description='stop data logging')
display(button2)

def stop_thread(c):
    global interrupt
    interrupt=True
    
button2.on_click(stop_thread)
```
